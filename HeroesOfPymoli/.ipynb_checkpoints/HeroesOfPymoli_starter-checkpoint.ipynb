## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins= [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999.9]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], age_bins, labels = age_groups)
purchase_data_df

purchase_data_agegroup = purchase_data_df.groupby("Age Group")
# #age_data

members = purchase_data_agegroup['SN'].nunique()
# # members

percentage_members = (members/players)*100

summary_df = pd.DataFrame ({"Total Counts": members, "Percentage OF Players":percentage_members })
summary_df

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins= [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999.9]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], age_bins, labels = age_groups)
purchase_data_df


purchase_data_agegroup = purchase_data_df.groupby("Age Group")
purchase_data_agegroup.count()

members = purchase_data_agegroup['SN'].count()
members

unique_members =purchase_data_agegroup['SN'].nunique()

average_price = purchase_data_agegroup["Price"].mean().map('{:.2f}'.format)
average_price

total_price = purchase_data_agegroup["Price"].sum()
total_price

average_per_person = (total_price/unique_members).map('{:.2f}'.format)
average_per_person

summary_df = pd.DataFrame ({"Purchase Counts": members, "Average Purchase Price":average_price, 
                            "Total Purchase Value":total_price, "Avg Total Purchase per Person":average_per_person })
summary_df
```
